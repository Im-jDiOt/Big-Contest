```python
import pandas as pd
import numpy as np
# 파일 경로
file1 = "C:/Users/User/Desktop/빅콘테스트/big_data_set1_f.csv"   # 기본정보 (업종/상권/개설일/폐업일 등)
file2 = "C:/Users/User/Desktop/빅콘테스트/big_data_set2_f.csv"   # 매출·운영 지표
file3 = "C:/Users/User/Desktop/빅콘테스트/big_data_set3_f.csv"   # 고객구성 지표

# CSV 불러오기
df1 = pd.read_csv(file1, encoding="cp949")
df2 = pd.read_csv(file2, encoding="cp949")
df3 = pd.read_csv(file3, encoding="cp949")

# ---- 컬럼명 통일 (공통키) ----
# big_data_set1에는 'TA_YM'이 없음 → 개설일 기준 등으로 생성 가능 (일단 None)
df1["TA_YM"] = None  

# big_data_set2, set3은 이미 ENCODED_MCT, TA_YM 있음

# ---- 중복 컬럼 정리 ----
# set2/set3 공통 키: ENCODED_MCT + TA_YM
# set1은 ENCODED_MCT만 있음
# → df1과 df2/df3 합칠 때는 ENCODED_MCT 기준 outer join, df2와 df3는 (ENCODED_MCT, TA_YM) 기준 outer join

# df2와 df3 먼저 병합
df23 = pd.merge(df2, df3, on=["ENCODED_MCT", "TA_YM"], how="outer")

# df1과 df23 병합 (ENCODED_MCT 기준)
final_df = pd.merge(df1, df23, on="ENCODED_MCT", how="outer")

# ---- 특수값 처리 (예: -999999.9, 'SV') ----
# 수치형 특수값 처리
num_cols = final_df.select_dtypes(include=["float64", "int64"]).columns
final_df[num_cols] = final_df[num_cols].replace(-999999.9, pd.NA)

# 문자열 특수값 처리
obj_cols = final_df.select_dtypes(include=["object"]).columns
final_df[obj_cols] = final_df[obj_cols].replace("SV", pd.NA)

# 결과 저장
final_df.to_csv("C:/Users/User/Desktop/빅콘테스트/whole_dataset.csv", index=False, encoding="utf-8-sig")

print("✅ 최종 병합 완료:", final_df.shape)
print("샘플 확인:")
print(final_df.head())
```

```python
import pandas as pd

# 파일 경로
file1 = "C:/Users/User/Desktop/빅콘테스트/big_data_set1_f.csv"   # 가맹점 기본정보
file2 = "C:/Users/User/Desktop/빅콘테스트/big_data_set2_f.csv"   # 운영/매출 지표
file3 = "C:/Users/User/Desktop/빅콘테스트/big_data_set3_f.csv"   # 고객구성 지표

# CSV 읽기 (한글 인코딩 문제 해결)
df1 = pd.read_csv(file1, encoding="cp949")
df2 = pd.read_csv(file2, encoding="cp949")
df3 = pd.read_csv(file3, encoding="cp949")

print("df1:", df1.shape)
print("df2:", df2.shape)
print("df3:", df3.shape)

# --------------------------------------------------------
# 1. df2 + df3를 ["ENCODED_MCT", "TA_YM"] 기준으로 merge
# --------------------------------------------------------
df23 = pd.merge(df2, df3, on=["ENCODED_MCT", "TA_YM"], how="outer")

print("df23 (df2+df3):", df23.shape)

# TA_YM이 일치하지 않는 경우 확인
missing_in_df3 = df23[df23["TA_YM"].isna()]
if not missing_in_df3.empty:
    print("⚠️ df2에는 있는데 df3에는 없는 (ENCODED_MCT, TA_YM) 조합 있음")
    print(missing_in_df3.head())

# --------------------------------------------------------
# 2. df1을 ["ENCODED_MCT"] 기준으로 merge
# --------------------------------------------------------
merged = pd.merge(df1, df23, on="ENCODED_MCT", how="outer")

print("최종 merged:", merged.shape)

# 불필요한 중복 컬럼 정리 (만약 생기면)
merged = merged.loc[:, ~merged.columns.str.contains("^Unnamed")]

# 저장
merged.to_csv("C:/Users/User/Desktop/빅콘테스트/merged_dataset.csv", index=False, encoding="utf-8-sig")
print("✅ 최종 병합 완료 & 저장")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. 데이터 불러오기
# -----------------------------
df = pd.read_csv("C:/Users/User/Desktop/빅콘테스트/merged_dataset.csv")

# -----------------------------
# 2. 결측치 확인
# -----------------------------
print("✅ 전체 결측치 개수:", df.isna().sum().sum())
print("\n📊 결측치 분포 (상위 20개):")
print(df.isna().sum().sort_values(ascending=False).head(20))

# 결측치 비율 시각화
missing_ratio = df.isna().mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=missing_ratio.head(15).index, y=missing_ratio.head(15).values)
plt.xticks(rotation=45)
plt.title("결측치 비율 (상위 15개 컬럼)")
plt.show()

# -----------------------------
# 3. 기본 통계치 확인 (이상치 탐지 힌트)
# -----------------------------
desc = df.describe(include="all")
print("\n📈 데이터 요약 통계:")
print(desc)

# -----------------------------
# 4. 수치형 컬럼 이상치 탐지 (IQR 기준)
# -----------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    if outliers > 0:
        print(f"⚠️ 컬럼 [{col}] 이상치 개수: {outliers}")

# -----------------------------
# 5. 처리 전략 추천 함수
# -----------------------------
def recommend_strategy(df):
    strategies = {}
    for col in df.columns:
        missing_ratio = df[col].isna().mean()
        dtype = df[col].dtype
        
        if missing_ratio > 0:
            if dtype == "object":
                strategies[col] = "카테고리 → 최빈값 대체 / '미기록' 추가"
            else:
                strategies[col] = "수치형 → 중앙값 대체 (또는 시계열이면 보간)"
        else:
            strategies[col] = "결측 없음"
    return pd.DataFrame.from_dict(strategies, orient="index", columns=["추천 처리 전략"])

print("\n📌 컬럼별 처리 전략 추천:")
print(recommend_strategy(df).head(20))
```
