```python
import pandas as pd
import numpy as np
# ÌååÏùº Í≤ΩÎ°ú
file1 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set1_f.csv"   # Í∏∞Î≥∏Ï†ïÎ≥¥ (ÏóÖÏ¢Ö/ÏÉÅÍ∂å/Í∞úÏÑ§Ïùº/ÌèêÏóÖÏùº Îì±)
file2 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set2_f.csv"   # Îß§Ï∂ú¬∑Ïö¥ÏòÅ ÏßÄÌëú
file3 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set3_f.csv"   # Í≥†Í∞ùÍµ¨ÏÑ± ÏßÄÌëú

# CSV Î∂àÎü¨Ïò§Í∏∞
df1 = pd.read_csv(file1, encoding="cp949")
df2 = pd.read_csv(file2, encoding="cp949")
df3 = pd.read_csv(file3, encoding="cp949")

# ---- Ïª¨ÎüºÎ™Ö ÌÜµÏùº (Í≥µÌÜµÌÇ§) ----
# big_data_set1ÏóêÎäî 'TA_YM'Ïù¥ ÏóÜÏùå ‚Üí Í∞úÏÑ§Ïùº Í∏∞Ï§Ä Îì±ÏúºÎ°ú ÏÉùÏÑ± Í∞ÄÎä• (ÏùºÎã® None)
df1["TA_YM"] = None  

# big_data_set2, set3ÏùÄ Ïù¥ÎØ∏ ENCODED_MCT, TA_YM ÏûàÏùå

# ---- Ï§ëÎ≥µ Ïª¨Îüº Ï†ïÎ¶¨ ----
# set2/set3 Í≥µÌÜµ ÌÇ§: ENCODED_MCT + TA_YM
# set1ÏùÄ ENCODED_MCTÎßå ÏûàÏùå
# ‚Üí df1Í≥º df2/df3 Ìï©Ïπ† ÎïåÎäî ENCODED_MCT Í∏∞Ï§Ä outer join, df2ÏôÄ df3Îäî (ENCODED_MCT, TA_YM) Í∏∞Ï§Ä outer join

# df2ÏôÄ df3 Î®ºÏ†Ä Î≥ëÌï©
df23 = pd.merge(df2, df3, on=["ENCODED_MCT", "TA_YM"], how="outer")

# df1Í≥º df23 Î≥ëÌï© (ENCODED_MCT Í∏∞Ï§Ä)
final_df = pd.merge(df1, df23, on="ENCODED_MCT", how="outer")

# ---- ÌäπÏàòÍ∞í Ï≤òÎ¶¨ (Ïòà: -999999.9, 'SV') ----
# ÏàòÏπòÌòï ÌäπÏàòÍ∞í Ï≤òÎ¶¨
num_cols = final_df.select_dtypes(include=["float64", "int64"]).columns
final_df[num_cols] = final_df[num_cols].replace(-999999.9, pd.NA)

# Î¨∏ÏûêÏó¥ ÌäπÏàòÍ∞í Ï≤òÎ¶¨
obj_cols = final_df.select_dtypes(include=["object"]).columns
final_df[obj_cols] = final_df[obj_cols].replace("SV", pd.NA)

# Í≤∞Í≥º Ï†ÄÏû•
final_df.to_csv("C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/whole_dataset.csv", index=False, encoding="utf-8-sig")

print("‚úÖ ÏµúÏ¢Ö Î≥ëÌï© ÏôÑÎ£å:", final_df.shape)
print("ÏÉòÌîå ÌôïÏù∏:")
print(final_df.head())
```

```python
import pandas as pd

# ÌååÏùº Í≤ΩÎ°ú
file1 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set1_f.csv"   # Í∞ÄÎßπÏ†ê Í∏∞Î≥∏Ï†ïÎ≥¥
file2 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set2_f.csv"   # Ïö¥ÏòÅ/Îß§Ï∂ú ÏßÄÌëú
file3 = "C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/big_data_set3_f.csv"   # Í≥†Í∞ùÍµ¨ÏÑ± ÏßÄÌëú

# CSV ÏùΩÍ∏∞ (ÌïúÍ∏Ä Ïù∏ÏΩîÎî© Î¨∏Ï†ú Ìï¥Í≤∞)
df1 = pd.read_csv(file1, encoding="cp949")
df2 = pd.read_csv(file2, encoding="cp949")
df3 = pd.read_csv(file3, encoding="cp949")

print("df1:", df1.shape)
print("df2:", df2.shape)
print("df3:", df3.shape)

# --------------------------------------------------------
# 1. df2 + df3Î•º ["ENCODED_MCT", "TA_YM"] Í∏∞Ï§ÄÏúºÎ°ú merge
# --------------------------------------------------------
df23 = pd.merge(df2, df3, on=["ENCODED_MCT", "TA_YM"], how="outer")

print("df23 (df2+df3):", df23.shape)

# TA_YMÏù¥ ÏùºÏπòÌïòÏßÄ ÏïäÎäî Í≤ΩÏö∞ ÌôïÏù∏
missing_in_df3 = df23[df23["TA_YM"].isna()]
if not missing_in_df3.empty:
    print("‚ö†Ô∏è df2ÏóêÎäî ÏûàÎäîÎç∞ df3ÏóêÎäî ÏóÜÎäî (ENCODED_MCT, TA_YM) Ï°∞Ìï© ÏûàÏùå")
    print(missing_in_df3.head())

# --------------------------------------------------------
# 2. df1ÏùÑ ["ENCODED_MCT"] Í∏∞Ï§ÄÏúºÎ°ú merge
# --------------------------------------------------------
merged = pd.merge(df1, df23, on="ENCODED_MCT", how="outer")

print("ÏµúÏ¢Ö merged:", merged.shape)

# Î∂àÌïÑÏöîÌïú Ï§ëÎ≥µ Ïª¨Îüº Ï†ïÎ¶¨ (ÎßåÏïΩ ÏÉùÍ∏∞Î©¥)
merged = merged.loc[:, ~merged.columns.str.contains("^Unnamed")]

# Ï†ÄÏû•
merged.to_csv("C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/merged_dataset.csv", index=False, encoding="utf-8-sig")
print("‚úÖ ÏµúÏ¢Ö Î≥ëÌï© ÏôÑÎ£å & Ï†ÄÏû•")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
# -----------------------------
df = pd.read_csv("C:/Users/User/Desktop/ÎπÖÏΩòÌÖåÏä§Ìä∏/merged_dataset.csv")

# -----------------------------
# 2. Í≤∞Ï∏°Ïπò ÌôïÏù∏
# -----------------------------
print("‚úÖ Ï†ÑÏ≤¥ Í≤∞Ï∏°Ïπò Í∞úÏàò:", df.isna().sum().sum())
print("\nüìä Í≤∞Ï∏°Ïπò Î∂ÑÌè¨ (ÏÉÅÏúÑ 20Í∞ú):")
print(df.isna().sum().sort_values(ascending=False).head(20))

# Í≤∞Ï∏°Ïπò ÎπÑÏú® ÏãúÍ∞ÅÌôî
missing_ratio = df.isna().mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=missing_ratio.head(15).index, y=missing_ratio.head(15).values)
plt.xticks(rotation=45)
plt.title("Í≤∞Ï∏°Ïπò ÎπÑÏú® (ÏÉÅÏúÑ 15Í∞ú Ïª¨Îüº)")
plt.show()

# -----------------------------
# 3. Í∏∞Î≥∏ ÌÜµÍ≥ÑÏπò ÌôïÏù∏ (Ïù¥ÏÉÅÏπò ÌÉêÏßÄ ÌûåÌä∏)
# -----------------------------
desc = df.describe(include="all")
print("\nüìà Îç∞Ïù¥ÌÑ∞ ÏöîÏïΩ ÌÜµÍ≥Ñ:")
print(desc)

# -----------------------------
# 4. ÏàòÏπòÌòï Ïª¨Îüº Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR Í∏∞Ï§Ä)
# -----------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    if outliers > 0:
        print(f"‚ö†Ô∏è Ïª¨Îüº [{col}] Ïù¥ÏÉÅÏπò Í∞úÏàò: {outliers}")

# -----------------------------
# 5. Ï≤òÎ¶¨ Ï†ÑÎûµ Ï∂îÏ≤ú Ìï®Ïàò
# -----------------------------
def recommend_strategy(df):
    strategies = {}
    for col in df.columns:
        missing_ratio = df[col].isna().mean()
        dtype = df[col].dtype
        
        if missing_ratio > 0:
            if dtype == "object":
                strategies[col] = "Ïπ¥ÌÖåÍ≥†Î¶¨ ‚Üí ÏµúÎπàÍ∞í ÎåÄÏ≤¥ / 'ÎØ∏Í∏∞Î°ù' Ï∂îÍ∞Ä"
            else:
                strategies[col] = "ÏàòÏπòÌòï ‚Üí Ï§ëÏïôÍ∞í ÎåÄÏ≤¥ (ÎòêÎäî ÏãúÍ≥ÑÏó¥Ïù¥Î©¥ Î≥¥Í∞Ñ)"
        else:
            strategies[col] = "Í≤∞Ï∏° ÏóÜÏùå"
    return pd.DataFrame.from_dict(strategies, orient="index", columns=["Ï∂îÏ≤ú Ï≤òÎ¶¨ Ï†ÑÎûµ"])

print("\nüìå Ïª¨ÎüºÎ≥Ñ Ï≤òÎ¶¨ Ï†ÑÎûµ Ï∂îÏ≤ú:")
print(recommend_strategy(df).head(20))
```
