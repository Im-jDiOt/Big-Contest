# 컬럼의 유니크 값을 통해 데이터 파악
- 서비스 업종 코드는 100대 생홟밀접 업종코드로 cs1~~1부터 9까지가 외식 대분류에 속함. 이 데이터들만 처리하면 됨.
  - data1의 업종 피쳐를 서비스 업종 코드로 통일할 필요가 있음.
- data1과 상권은 x,y 좌표 값을 바탕으로 연결할 예정.

```python
def get_unique_of_col(df):
	for col in df.columns:
		print("==", col, "==")
		print(df[col].unique())
```
