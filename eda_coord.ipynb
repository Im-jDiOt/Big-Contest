```python
max_x, min_x = max(data1['X']), min(data1['X'])
max_y, min_y = max(data1['Y']), min(data1['Y'])

filtered_sg_coord = sg_coord[
    sg_coord["엑스좌표_값"].between(min_x - 1000, max_x + 1000) &
    sg_coord["와이좌표_값"].between(min_y - 1000, max_y + 1000)
]
```

```python
data1.merge(data2, how="left", on="ENCODED_MCT")
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

# data1 데이터 세트 그리기 (파란색 원)
ax.scatter(data1['X'], data1['Y'], label='Data1', color='blue', alpha=0.7, s=50)

# sg_coord 데이터 세트 그리기 (빨간색 X)
ax.scatter(filtered_sg_coord['엑스좌표_값'], filtered_sg_coord['와이좌표_값'], label='Sg_coord', color='red', marker='x', s=50)

# 제목, 축 라벨 및 범례 추가
ax.set_title('Scatter Plot of Data1 and Sg_coord')
ax.set_xlabel('X Coordinate (엑스좌표_값)')
ax.set_ylabel('Y Coordinate (와이좌표_값)')
ax.legend()
ax.grid(True)

# 플롯 저장
plt.savefig('data\\fig\\combined_scatter_plot.png')
```
